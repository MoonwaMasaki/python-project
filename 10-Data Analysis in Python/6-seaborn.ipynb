## Data Visualization with Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualization with just a few lines of code.

```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn 
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

```python
# Line Plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line plot of Size vs Total Bill")
plt.show()
```

```python
# Categorical Plots
# Bar plot 

sns.barplot(x='day', y='total_bill', data = tips )
plt.title("Bar Plot of Day vs Total Bill")
plt.show()
```

```python
# Box Plot 
sns.boxplot(x= 'day', y='total_bill', data = tips)
plt.title("Box Plot")
plt.show()
```

```python
# Violin Plot 

sns.violinplot(x='day', y = 'total_bill', data = tips)
plt.title("Violin Plot")
plt.show()
```

```python
# Histogram 

sns.histplot(tips['total_bill'], bins=10,kde= True)
```

```python
# KDE Plot 
sns.kdeplot(tips['total_bill'], fill= True)
```

```python
# Pair Plot

sns.pairplot(tips)
```

```python
# HEATmap
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot= True, cmap='coolwarm')
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# Total Sales by Product 

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum)
plt.title("Total Sales by Product")
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Total Sales by Region

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator=sum)
plt.title("Total Sales by Region")
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
